# Triangle, Pentagonal, and Hexagonal Numbers
## Project Euler Problem 45
Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:
<br/>
Triangle	 	$T_n=\frac{n(n+1)}{2}$	 	1, 3, 6, 10, 15, ...<br/>
Pentagonal	 	$P_n=\frac{n(3n−1)}{2}$	 	1, 5, 12, 22, 35, ...<br/>
Hexagonal	 	$H_n=n(2n−1)$	 	1, 6, 15, 28, 45, ...<br/>
It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.

Find the next triangle number that is also pentagonal and hexagonal.

```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
n_vec = list(range(10000,100000,1))
nums = pd.DataFrame(index=n_vec, columns=['Triangular', 'Pentagonal', 'Hexagonal'])
for n in n_vec:
    nums.loc[n,'Triangular'] = (n*(n+1))/2
    nums.loc[n, 'Pentagonal'] = (n*((3*n)-1))/2
    nums.loc[n,'Hexagonal'] = n*((2*n)-1)
    
nums.head()
```

```python
reduce(np.intersect1d, [nums['Triangular'], nums['Pentagonal'], nums['Hexagonal']])
```
